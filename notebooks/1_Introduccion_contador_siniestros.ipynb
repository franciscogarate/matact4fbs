```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('matact4fbs/data/siniestros.csv', sep=';')
df
```

```python
df.describe()
```

```python
df_contador = pd.crosstab(df.Poliza, columns='contador')
df_contador
```

```python
df_freq = pd.crosstab(df_contador.contador, 'frecuencia')
df_freq
```

```python
plt.pie(df_freq.frecuencia, labels=df_freq.index, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
plt.bar(df_freq.index, df_freq.frecuencia)
plt.show()
```
