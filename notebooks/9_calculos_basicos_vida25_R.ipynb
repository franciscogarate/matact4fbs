```r
#Calculemos los lx, los metemos en un vector columna (tenemos en cuenta el desfase)
L<-numeric()
L[1]<- 1000000*(1-pasem1[1])
```

```r
for (i in 2:113) { L[i] <- L[i-1]*(1-pasem1[i])}
L
plot(L)
plot(pasem1)
```

```r
#Probabilidades basicas
px <- function(x) {L[x+1]/L[x]}
px(20)
```

```r
qx <- function(x) {1-(L[x+1]/L[x])}
qx(20)
```

```r
tpx <- function(x,t) {L[x+t]/L[x]}
tpx(20,10);tpx(50,40)
```

```r
tqx <- function(x,t) {1-(L[x+t]/L[x])}
tqx(20,10)
```

```r
kqx<- function(x,t) {(L[x+t]-L[x+t+1])/L[x]}
kqx(20,10)
```

```r
#Factor de actualizacion actuarial
tEX<-function(x,n,i) {(1+i)^(-n)*L[x+n]/L[x]}
tEX(20,50,0.03)
```

```r
#valor actual actuarial del seguro vida entera
Ax<-function(x,i) {t<-c(0:(113-x-1));sum((1+i)^(-t-1)*((L[x+t]-L[x+t+1])/L[x]))}
Ax(30,0.02)
```

```r
#valor actual actuarial del seguro temporal
Axn<-function(x,n,i) {t<-c(0:(n-1));sum((1+i)^(-t-1)*((L[x+t]-L[x+t+1])/L[x]))}
Axn(30,10,0.02)
```

```r
#renta vitalicia prepagable
ax<-function(x,i) {t<-c(0:(113-x-1));sum((1+i)^(-t)*(L[x+t]/L[x]))}
ax(40,0.03)
ax(40,0.02)
```

```r
#renta temporal prepagable
axn<-function(x,n,i) {t<-c(0:(n-1));sum((1+i)^(-t)*(L[x+t]/L[x]))}
axn(40,10,0.03)
```

```r
#renta diferida nax=prima unica
nax<-function(x,n,i){tEX(x,n,i)*ax(x+n,i)}
nax(20,10,0.03);nax(20,10,0.02);nax(20,45,0.03);nax(20,45,0.02);
```

```r
#renta vitalicia variable en progresion geometrica
qax<-function(x,i,q) {t<-c(0:(113-x-1));sum((q^(t))*(1+i)^(-t)*(L[x+t]/L[x]))}
qax(40,0.03,1)
```

```r
#renta temporal variable en progresion geometrica
qaxn<-function(x,n,i,q) {t<-c(0:(n-1));sum((q^(t))*(1+i)^(-t)*(L[x+t]/L[x]))}
qaxn(40,10,0.03,1)
qaxn(40,10,0.03,1.05)
```
