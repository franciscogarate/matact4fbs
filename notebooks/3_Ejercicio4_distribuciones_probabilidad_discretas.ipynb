```python
import pandas as pd
import numpy as np
from scipy.stats import poisson

df = pd.DataFrame({
    'numero': [0, 1, 2, 3, 4],
    'contador': [37042, 4654, 398, 32, 1]
})
```

```python
total = df['contador'].sum()
lambda_param = (df['numero'] * df['contador']).sum() / total

print(f"λ estimado: {lambda_param:.4f}")
print(f"Total de observaciones: {total}")

# Calcular probabilidades teóricas de Poisson
df['prob_teorica'] = poisson.pmf(df['numero'], lambda_param)

# Calcular contadores teóricos
df['contador_teorico'] = df['prob_teorica'] * total
df
```

```python
# Calcular chi-cuadrado para bondad de ajuste
chi2 = ((df['contador'] - df['contador_teorico'])**2 / df['contador_teorico']).sum()
print(f'χ² = {chi2}')
```

### Binomial negativa

```python
from scipy.stats import nbinom

df = pd.DataFrame({
    'numero': [0, 1, 2, 3, 4],
    'contador': [37042, 4654, 398, 32, 1]
})

total = df['contador'].sum()
media = (df['numero'] * df['contador']).sum() / total
varianza = ((df['numero'] - media)**2 * df['contador']).sum() / total

print(f"Media: {media:.4f}")
print(f"Varianza: {varianza:.4f}")
print(f"Ratio Varianza/Media: {varianza/media:.4f}")

# Estimar parámetros de la binomial negativa (método de momentos)
# Para scipy: media = n(1-p)/p, varianza = n(1-p)/p²
p = media / varianza
n = media * p / (1 - p)

print(f"\nParámetros estimados:")
print(f"n = {n:.4f}")
print(f"p = {p:.4f}")

df['prob_teorica'] = nbinom.pmf(df['numero'], n, p)
df['contador_teorico'] = df['prob_teorica'] * total
df
```

```python
# Chi-cuadrado para bondad de ajuste
chi2 = ((df['contador'] - df['contador_teorico'])**2 / df['contador_teorico']).sum()
print(f'χ² = {chi2}')
```
